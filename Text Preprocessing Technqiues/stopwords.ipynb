```python
## Speech Of DR APJ Abdul Kalam
paragraph = """I have three visions for India. In 3000 years of our history, people from all over 
               the world have come and invaded us, captured our lands, conquered our minds. 
               From Alexander onwards, the Greeks, the Turks, the Moguls, the Portuguese, the British,
               the French, the Dutch, all of them came and looted us, took over what was ours. 
               Yet we have not done this to any other nation. We have not conquered anyone. 
               We have not grabbed their land, their culture, 
               their history and tried to enforce our way of life on them. 
               Why? Because we respect the freedom of others.That is why my 
               first vision is that of freedom. I believe that India got its first vision of 
               this in 1857, when we started the War of Independence. It is this freedom that
               we must protect and nurture and build on. If we are not free, no one will respect us.
               My second vision for India’s development. For fifty years we have been a developing nation.
               It is time we see ourselves as a developed nation. We are among the top 5 nations of the world
               in terms of GDP. We have a 10 percent growth rate in most areas. Our poverty levels are falling.
               Our achievements are being globally recognised today. Yet we lack the self-confidence to
               see ourselves as a developed nation, self-reliant and self-assured. Isn’t this incorrect?
               I have a third vision. India must stand up to the world. Because I believe that unless India 
               stands up to the world, no one will respect us. Only strength respects strength. We must be 
               strong not only as a military power but also as an economic power. Both must go hand-in-hand. 
               My good fortune was to have worked with three great minds. Dr. Vikram Sarabhai of the Dept. of 
               space, Professor Satish Dhawan, who succeeded him and Dr. Brahm Prakash, father of nuclear material.
               I was lucky to have worked with all three of them closely and consider this the great opportunity of my life. 
               I see four milestones in my career"""
```

#### Reason why we are using Stop words

As we can see in the above paragraph or corpus there are some words like i, the , have, you ,he, she ,we ,and,of to ,they so all these words does not play important role when we do task like spam classification or whether this is a positive or negative review so by using stop words we are going to remove the un important words because this kind of use cases where you specifically focusing the important words to determine the output, so lets pass this entire corpus to the stop words and lets see what happens

##### Along with Stop words we can import stemming

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
# and i have to download the stopwords in order to see the stopwords in differernt langugages, by this below code we can able to download the different langauges stopwords which are already available nltk library
import nltk
nltk.download('stopwords')
```

```python
x=stopwords.words('english')  # lets check which are all stop words are available in the english language and other than the available stop words which are available in english language we can also able to add the customize stopwords and we can remove that customize stopwords in the paragraph at once
x
```

```python
stopwords.words('arabic') # these are list of available stopwords for german langauge if we are working on german paragaph then with the help of the exsiting german lngauge stop words we can able to remove these all stop words in the paragraph
```

Now iam going to perform 2 things
1) performing the Tokenizer on all sentences as it results all sentences  into words
2) converting all the sentences to lower() 
3) Performing the Stopwords
4) Performing the Stemming,snowball lemmatization on non stopwords

```python
sentences=nltk.sent_tokenize(paragraph)  # first i performed the tokenizer on the corpus or paragraph it means i converted the whole paragraph into list of sentences 
```

```python
type(sentences)
```

```python
# Convert each sentence to lowercase using list comprehension  because in our corpus we may some words which get repeats in smaller case and upper case so then our model cannot able to find the stop words for the uppercase words because by default the set of stopwords in eglish langauge is defined in lower case so thats y we convert all words into lower case and then we can easily find the which all words are stopword
sentences = [sentence.lower() for sentence in sentences]
sentences
```

### Applying Stopwords on sentences
#### Now iam going to traverse all the sentences and apply the stop words on those tokenizer corpus which is sentences then i will find which words are not stop from the sentences i will take those words which are not stopwords and then i apply stemming in the non stopwords

```python

from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
# apply stopwords and filter and then apply stemming on the non stopwords
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i]) # here iam furthure performing the tokenization on the sentences then as a result we will get words from each sentence
    words=[stemmer.stem(word) for word in words if word not in set(x)]  # here by using list comprehension i written logic to find the root word on the non stopwords means each words are going to traivers each word which we get from each sentence if the words falls under the set of stopwords of english langauge we remove those and which ever the words does not falls under the stop words we apply stemming on those non stop words and i mention the set of stopwords of particular langauge here in this case we need to use english langugae set of stopwords because our coprus is present in english langauge
    sentences[i]=' '.join(words) # and after find the non stop words iam going to return those all nonstop words as sentences
```

#### Now iam going to apply the lemmatization on the nonstopwords to get good gramitical form of root words or lemmas 

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
## Apply Stopwords And Filter And then Apply Snowball Stemming

for i in range(len(sentences)):
    #sentences[i]=sentences[i].lower()
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word.lower(),pos='v') for word in words if word not in set(x)]  # here most of the words in the paragraph are verbs thats why i taken post tag as v which stands for verb
    sentences[i]=' '.join(words)# converting all the list of words into sentences
```
