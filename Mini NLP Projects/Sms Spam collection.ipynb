#### Data cleaning and Preprocessing by NLP Techniques

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(message)):
    review=re.sub('[^a-zA-Z0-9]',' ',message['messages'][i])
    review=review.lower() # this line helps us to prevent the duplicate texts
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

review=re.sub('[^a-zA-z]',' ',message['messages'][i]): This line removes any characters that are not alphabetic letters from the text message. The regular expression [^a-zA-z] matches any character that is not an alphabetic letter, and re.sub() replaces those characters with a space ' '.

review=review.lower(): This line converts all the alphabetic characters in the text message to lowercase. This ensures that the text is case-insensitive and helps in preventing duplicate texts due to case variations.

review=review.split(): This line splits the text message into a list of individual words. By default, it splits the text wherever it encounters whitespace characters (spaces, tabs, newlines, etc.), effectively separating the text into individual words.

review=[ps.stem(word) for word in review if not word in stopwords.words('english')]: This line performs stemming using the PorterStemmer (ps) from the NLTK library on each word in the list review. Stemming reduces words to their root or base form, which can help in reducing the dimensionality of the feature space. Additionally, it removes any stopwords (commonly occurring words that typically do not carry significant meaning) from the list of words.

review=' '.join(review): Finally, this line joins the list of stemmed words back into a single string, with each word separated by a space. This reconstructed string is then appended to the corpus list.

### Create Bag Of Words

Conveting the Text to vectors inorder to make our model to understand about the data which we will train furthur

```python
## Create the Bag OF Words model
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv=CountVectorizer(max_features=2500,ngram_range=(1,2),binary=True) # we are selecting the max features 2500 based on there frequency of repeatation of same word and i have taken n gram where i taken min as 1 and max value as 2  means first will select the single feature and then once 1 done with selecting the single feature then in sequence i collect the 2 features at same time till last of the corpus it means that you are generating both unigrams (single words) and bigrams (sequences of two adjacent words) from your corpus.,binary=True helps us to maintain the assigin the presence of particular word in the document irrespective even the word repeated for many times still it maintains to show the values between 0 to 1 
X=cv.fit_transform(corpus).toarray() # here finally i converted to array of those converted vectors 
```

```python
X.shape  # as we can see there are total top 2500 features we selected based on the no of times each feature got repeated and there are 5572 documents or rows 
```

```python
y=pd.get_dummies(message['labels'])
print(y) # here i have used the onehot encoding and i have created the 2 columns for ham and spam with the help of one hot encoding by using this get_dummis
y=y.iloc[:,0].values # here we are selecting the all the rows and only the first colum which is ham and storing it in y variable
print()
```

```python
print(y)
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

```python
X_train,y_train
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python

spam_detect_model=MultinomialNB().fit(X_train,y_train)
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python

from sklearn.metrics import accuracy_score,classification_report
```

```python

accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

####  Now performing the word Embedded using Tf-Idf technique to convert words to vectors

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(corpus,y,test_size=0.20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
X_train=tv.fit_transform(X_train).toarray()
X_test=tv.transform(X_test).toarray()
```

```python
X_train,y_train
```

```python
tv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_tfidf_model = MultinomialNB().fit(X_train, y_train)
```

```python

#prediction
y_pred=spam_tfidf_model.predict(X_test)
```

```python

score=accuracy_score(y_test,y_pred)
print(score)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```
